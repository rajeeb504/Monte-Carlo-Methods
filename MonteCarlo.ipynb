Monte Carlo simulation to find the expected number of fair coin tosses to get n consecutive heads.

```python
import random

def Expectation_nH(n, N= 100000):

  s = ['H', 'T']
  total = 0

  for _ in range(N):
    toss = 0
    heads = 0
    while heads < n:
      toss += 1
      if random.choice(s) == 'H':
        heads += 1
      else:
        heads = 0

    total += toss

  result = total / N

  return result
```

Run an experiment to get 3 consecutive heads

```python
e = Expectation_nH(3)

print(f"Expected number of tosses to get 3 consecutive heads is: {e}")
```
